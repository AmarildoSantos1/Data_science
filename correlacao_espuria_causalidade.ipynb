```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('spotify_top_1000_tracks.csv')

# Converter a coluna de data
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Configuraes de estilo
sns.set(style="whitegrid")

# 1. Correlao: Durao vs Popularidade
plt.figure(figsize=(8,6))
sns.scatterplot(x='duration_min', y='popularity', data=df)
plt.title('Correlao entre Durao da Msica e Popularidade')
plt.xlabel('Durao (minutos)')
plt.ylabel('Popularidade')
plt.show()

# 2. Causalidade: Durao influencia Popularidade (hiptese)
plt.figure(figsize=(8,6))
sns.regplot(x='duration_min', y='popularity', data=df, scatter_kws={'alpha':0.5})
plt.title('Causalidade: Durao da Msica influencia Popularidade')
plt.xlabel('Durao (minutos)')
plt.ylabel('Popularidade')
plt.show()

# 3. Correlao Espria: Data de Lanamento vs Popularidade
plt.figure(figsize=(10,6))
sns.scatterplot(x='release_date', y='popularity', data=df)
plt.title('Popularidade vs. Data de Lanamento (Correlao Espria)')
plt.xlabel('Data de Lanamento')
plt.ylabel('Popularidade')
plt.show()
```

### Correlao: Durao e popularidade esto correlacionadas de forma leve.

### Causalidade: A durao das msicas afeta sua aceitao no mercado e, consequentemente, sua popularidade (existe uma faixa ideal de durao).

### Correlao espria: Msicas lanadas recentemente tendem a ser menos populares, no por serem piores, mas por estarem h menos tempo na plataforma.
